# Problem 2: Data Augmentation and APIs

Auxiliary functions for analysis

```python
def prepare_database(df):
    '''
    Add empty columns of the variables of interest to be
    filled by the API
    '''
    df["income"] = np.nan
    df["median_age"] = np.nan
    df["median_rent"] = np.nan
    df["population"] = np.nan
    df["whites"] = np.nan
    df["blacks"] = np.nan

    return df
```

```python
def get_response(zip_code):
    '''
    Make the query to the API and return a list
    '''
    var = 'B07011_001E,B01002_001E,B25064_001E,B01001_001E,B01001A_001E,B01001B_001E' 
    census_key = ''
    target = "http://api.census.gov/data/2015/acs5?get={}&for=zip+code+tabulation+area:{}&key={}"
    r = requests.get(target.format(var, zip_code, census_key))
    response = json.loads(r.text)
    return response[1]
```

```python
def set_values(df):
    '''
    Create a dataframe with the relevant variables. Only one value by
    ZIP Code
    '''
    zip_codes = list(set(df['ZIP Code']))
    dictionary = {'ZIP Code': zip_codes}
    aux_df = pd.DataFrame(dictionary)
    iterative_df = prepare_database(aux_df)

    for i, row in iterative_df.iterrows():
        zip_code = str(int(row['ZIP Code']))
        try: 
            response = get_response(zip_code)
        except:
            response = ['','','','','','']

        iterative_df.set_value(i,'income',response[0])
        iterative_df.set_value(i,'median_age',response[1])
        iterative_df.set_value(i,'median_rent',response[2])
        iterative_df.set_value(i,'population',response[3])
        iterative_df.set_value(i,'whites',response[4])
        iterative_df.set_value(i,'blacks',response[5])
        
        try:
            iterative_df["p_whites"] = iterative_df["whites"]/iterative_df["population"]
            iterative_df["p_blacks"] = iterative_df["blacks"]/iterative_df["population"] 
        except:
            None

    return iterative_df
```

```python
def create_final_db(df):
    '''
    Merge the dataframe of relevant variables from the API with the original df
    '''
    iterative_df = set_values(df)
    rv = pd.merge(df, iterative_df, left_on = "ZIP Code", right_on = "ZIP Code")
    return rv
```

## Report for a policy audience

The graffiti removal reports received by the 311 system in the City of Chicago varies depending on the ethnicity of people who lives in different zip codes areas. For instance, in places where the percentage of whites is very high, the locations where the graffiti is reported are express ways, overpass, mailbox, and dumpsters. Meanwhile,   in places with a higher percentage of blacks, the most reported locations are viaduct, garage, bench, and phone. 

Regarding the sanitations reports, the kind of calls changes among different rent levels. For example, in places where the rent is high (potentially with many new building developments), the most frequent reports are construction site cleanliness, dumpsters not being emptied, and garbage in the alley. All of those are related to the value of the properties, and may reflect the importance of construction, and the importance to attract new people to the neighborhood. On the other hand, in places with a lower median rent, the priorities appear to be different. In those locations, the most reported issues are graffiti in commercial vehicles, garbage in yards, and dog fences in the yard. All those are more related to daily-life experiences, and maybe in places where residential buildings development is not as important as in areas with higher median rents.
